```python
from mpl_toolkits.mplot3d import Axes3D
from math import sin, cosh, pi, sinh
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
fig = plt.figure(figsize = (11,8))
ax = fig.gca(projection='3d')

X = np.arange(0, 1, 0.01)
Y = np.arange(0, 1, 0.01)
X, Y = np.meshgrid(X, Y)

def A(X,Y):
    P_1 = 300.
    P_2 = 150.
    P_3 = 100.
    return P_2 + sum([(P_3/pi)*((1-(-1)**n)/(n*np.sinh(pi*n)))*np.sinh(pi*n*X)*np.sin(pi*n*Y) + (P_1/pi)*((1-(-1)**n)/(n*np.sinh(pi*n)))*np.sinh(pi*n*(X-1))*np.sin(pi*n*Y)  for n in range(1,100)])

surf = ax.plot_surface(X, Y, A(X,Y), cmap=cm.coolwarm, edgecolor = 'black')
ax.set_xlabel('X', fontweight = 'bold')
ax.set_ylabel('Y', fontweight = 'bold')
ax.set_zlabel('Давление P', fontweight = 'bold')
ax.view_init(25, 120)
```

```python
def AA(x,y):
    P_1 = 300.
    P_2 = 150.
    P_3 = 100
    return P_2 + sum([(P_3/pi)*((1-(-1)**n)/(n*np.sinh(pi*n)))*np.sinh(pi*n*x)*np.sin(pi*n*y) + (P_1/pi)*((1-(-1)**n)/(n*np.sinh(pi*n)))*np.sinh(pi*n*(x-1))*np.sin(pi*n*y)  for n in range(1,100)])
```

```python
x = np.arange(0, 1, 0.01)
fig, ax = plt.subplots(2, 3, figsize=(12, 9))
axs = [ax[i, j] for i in range(2) for j in range(3)]
colors = iter(['red', 'blue', 'purple', 'green', 'black', 'darkred'])
y_values = iter(['y = 0', 'y = 0.2', 'y = 0.4', 'y = 0.6', 'y = 0.8', 'y = 1.0'])
for k,l in zip(np.arange(0, 1.2, 0.2), range(0,7)):
    axs[l].plot(x, AA(x,y = k), color = next(colors))
    axs[l].set_title(next(y_values), fontweight = 'bold' )
    axs[l].set_ylabel('P', rotation = 0, fontsize = 13, fontweight = 'bold')
    for j in range(3,6):
        axs[j].set_xlabel('x', fontsize = 13, fontweight = 'bold')
    axs[l].grid()
    axs[l].set_ylim(0,210)
```
